```python
gdb_path = 'data/canopy.gdb'
```

```python
layers = fiona.listlayers(gdb_path)
print(layers)
```

```python
gdf = gpd.read_file(gdb_path,layer=layers[0], engine='pyogrio')
print(gdf.head())
print(gdf.crs)
print(gdf.columns)
```

```python
gdf.plot(figsize=(8,6))
```

```python
m = gdf.explore(
    column="Shape_Area",          
    cmap="Greens",                
    tooltip=["Type", "Shape_Area"],  
    tiles="CartoDB positron",     
    height=600,
    width="100%"
)

m
```
